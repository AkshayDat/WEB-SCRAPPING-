```python
# Importing important Libraries
import pandas as pd
import selenium
from selenium import webdriver
from selenium.webdriver.chrome.options import Options   
from selenium.common.exceptions import InvalidSelectorException,NoSuchElementException
import time
from selenium .webdriver import ActionChains
chrome_options = Options()
chrome_options.add_argument('--incognito')
from selenium .webdriver.common.keys import Keys
```

```python
#Connecting to the web server
driver=webdriver.Chrome(r"C:/Users/Akshai/Downloads/chromedriver_win32 (1)/chromedriver.exe")

#specifying the web page
driver.get("https://www.deccanherald.com")

#scrolling down to the page at the end
driver.execute_script("window.scrollTo(0,document.body.scrollHeight)")   #studied this on Pythonbasics.org
time.sleep(3)

#openeing the sitemap of the deccanherald.com page
driver.find_element_by_xpath("//div[@class='newfooter_section_two_blocks_last']/ul[2]/li[4]/a").click()
```

```python
#switiching focus to the sitemap page
tab_1=driver.window_handles[1]         
driver.switch_to.window(tab_1)
```

```python
#crating list to store the Month click reference
mnth=[]
```

```python
#creating list to store scraped data
Date=[]
Author=[]
Vertical=[]
Headline=[]
Description=[]
```

```python
#defining function to scrape web_element

def myscrape():
    
    try:
        
        txt=driver.find_element_by_xpath("//ul[@class='crud-items']/li[1]").text
        a=txt.split(',')[0]
        #print(a)
        Date.append(a)
    except NoSuchElementException as e:
        Date.append("-")
        #print("------")
        
    try:    
        Verti=driver.find_element_by_xpath("//div[@class='item-list']/ul/li[2]/a").text
        Vertical.append(Verti)
        #print(Verti)
        
    except NoSuchElementException as e:
         Vertical.append("-")
         #print("------")

    try:
        
        Head=driver.find_element_by_xpath("//div[@class='strap-heading-wrapper']/h1").text
        Headline.append(Head)
        #print(Head)
        
    except NoSuchElementException as e:
        Headline.append("-")
        #print("------")


    try:
        
        Desp=driver.find_element_by_xpath("//div[@class='field field-name-body field-type-text-with-summary field-label-hidden']").text
        Description.append(Desp)
        #print(Desp)
        
        
        
    except NoSuchElementException as e:
        Description.append("-")
        #print("------")
    
    
    try:
        
        auth=driver.find_element_by_xpath("//div[@class='article-author__time']/span").text
        Author.append(auth)
        #print(auth)
    except NoSuchElementException as e:
        Author.append("-")
        #print("------")
    
    driver.close()
    
    
    
    #switiching focus to the new tab
    tab_1=driver.window_handles[3]         
    driver.switch_to.window(tab_1)
  
```

```python
#collecting the click element for september 2016 and december 2016
mnth.append(driver.find_element_by_xpath("//ul[@class='toplevel']/li[8]/ul/li[9]/a"))
mnth.append(driver.find_element_by_xpath("//ul[@class='toplevel']/li[8]/ul/li[12]/a"))
```

```python
#defining the function to store and perform buttton click and element click
def scrape():
    
    button=driver.find_elements_by_xpath("//ul[@class='taxonomy-tree children']/li/a")
    print(len(button))
    
    for btn in button[12:len(button)]:#len(button)
        j=2
        #btn.click()
        act=ActionChains(driver)
        act.context_click(btn).key_down(Keys.CONTROL).click(btn).perform()
    
        #switiching focus to the new tab
        tab_2=driver.window_handles[j+1]         
        driver.switch_to.window(tab_2)
    
        elem_=driver.find_elements_by_xpath("//li[@class='group sanspro-reg archives-note']/a")
        print(len(elem_))
    
        for i in elem_[0:len(elem_)]:
            k=3
            act=ActionChains(driver)
            act.context_click(i).key_down(Keys.CONTROL).click(i).perform()
            #switiching focus to the new tab
            tab_2=driver.window_handles[k+1]         
            driver.switch_to.window(tab_2)
            myscrape()
    
        driver.close()
    

        #checking the lenght of each list
        print(len(Date),len(Author),len(Vertical),len(Headline),len(Description))

        #switiching focus to 2016 oct storeies
        tab_1=driver.window_handles[2]         
        driver.switch_to.window(tab_1)
        
    driver.close()
        
    #switiching focus to 2016 oct storeies
    tab_1=driver.window_handles[1]         
    driver.switch_to.window(tab_1)
        
        
    
    
    
```

```python
#initializing for loop for scraping information
for month in mnth[1:2]:
        l=1
        act=ActionChains(driver)
        act.context_click(month).key_down(Keys.CONTROL).click(month).perform()
        #switiching focus to the new tab
        tab_2=driver.window_handles[l+1]         
        driver.switch_to.window(tab_2)
        scrape()
```
